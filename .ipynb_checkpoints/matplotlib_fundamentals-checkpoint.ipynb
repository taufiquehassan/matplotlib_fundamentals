```python
x = [1,2,3,4]
y = [11,22,33,44]
plt.plot(x,y)
plt.show()
```

```python
# 1st method
fig = plt.figure()
ax = fig.add_subplot()
plt.show()
```

```python
# 2nd method
fig = plt.figure()
ax = fig.add_axes([1,1,1,1])
ax.plot(x,y)
plt.show()
```

```python
# 3rd method (Recommend)
fig, ax = plt.subplots()
ax.plot(x,y)
plt.show()
```

### matplotlib workflow

```python
#0. import matplotlib
import matplotlib.pyplot as plt

#1. prepare data
x = [1,2,3,4,5]
y = [11,22,33,44,55]

#2. setup plot
fig, ax = plt.subplots(figsize = [10,10])

#3. plot data
ax.plot(x,y)

#4. customiz plot
ax.set(title = "simpole data",
       xlabel = "x-axis", 
       ylabel = "Yaxis")
      
#5. save and show
fig.savefig("sampleplot.png")
```

```python
#Bar Plot
nut_butter_prices = {"Almond Butter" : 10,
                    "Peanut Butter": 8,
                    "Cashew butter": 20}

fig, ax = plt.subplots()
ax.bar(nut_butter_prices.keys(), nut_butter_prices.values());
ax.set(title = "Taufiques BUtter Prices",
      ylabel = "Price in ($)")
```
