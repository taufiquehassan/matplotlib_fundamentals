### Which method should you use for plotting?
* Plotting something quickly? - use pyplot
* Plotting  something advanced? - use OO method

```python
over_50 = heart_disease[heart_disease["age"] > 50]
len(over_50)
```

```python
# pyplot method
over_50.plot(kind = "scatter",
             x = 'age',
             y = 'chol',
             c = 'target');
```

```python
# oo method from pyplot
fig, ax = plt. subplots( figsize = (10, 6))
over_50.plot(kind = "scatter",
             x = 'age',
             y = 'chol',
             c = 'target',
             ax=ax);
#ax.set_xlim([50,80])
```

```python
# oo method from scratch
fig, ax = plt. subplots( figsize = (10, 6))

scatter = ax.scatter(x = over_50['age'],
                     y = over_50['chol'],
                     c = over_50['target'],)

ax.set(title = "heart disease",
       xlabel = "age (year)",
       ylabel = "cholesterol")

#add legend
ax.legend(*scatter.legend_elements(), title = "Target")

#add horizontal line
ax.axhline(over_50['chol'].mean(),
           linestyle = 'dashed');
```

```python
#Subplot

fig, (ax0, ax1) = plt.subplots( nrows =2,
                        ncols = 1,
                        figsize = (10,10),
                         sharex = True)

scatter = ax0.scatter( x = over_50["age"],
                       y = over_50['chol'],
                       c = over_50['target'])
ax0.set(title = "cholsterol level",
        ylabel = "Cholesterol")
ax0.legend(*scatter.legend_elements(), title = 'target')
ax0.axhline(y = over_50["chol"].mean(),
            linestyle = '--')

scatter = ax1.scatter( x = over_50["age"],
                       y = over_50['thalach'],
                       c = over_50['target'])
ax1.set(title = "thalach level",
        xlabel = "age",
        ylabel = "thalach")
ax1.legend(*scatter.legend_elements(), title = 'target')
ax1.axhline(y = over_50["thalach"].mean(),
            linestyle = '--')

fig.suptitle('Heart Disease Analysis', fontsize=12, fontweight='bold')
```

```python
# oo method from scratch
plt.style.use('seaborn-whitegrid')
fig, ax = plt. subplots( figsize = (10, 6))

scatter = ax.scatter(x = over_50['age'],
                     y = over_50['chol'],
                     c = over_50['target'],
                     cmap= 'winter')  #this changes the color scheme

ax.set(title = "heart disease",
       xlabel = "age (year)",
       ylabel = "cholesterol")

#add legend
ax.legend(*scatter.legend_elements(), title = "Target")

#add horizontal line
ax.axhline(over_50['chol'].mean(),
           linestyle = 'dashed');
```

```python
#Subplot
plt.style.use('seaborn-whitegrid')
fig, (ax0, ax1) = plt.subplots( nrows =2,
                        ncols = 1,
                        figsize = (10,10),
                        sharex = True)

scatter = ax0.scatter( x = over_50["age"],
                       y = over_50['chol'],
                       c = over_50['target'],
                       cmap= 'winter')
ax0.set(title = "cholsterol level",
        ylabel = "Cholesterol",
        ylim = [100,600])

ax0.legend(*scatter.legend_elements(), title = 'target')
ax0.axhline(y = over_50["chol"].mean(),
            linestyle = '--')

scatter = ax1.scatter( x = over_50["age"],
                       y = over_50['thalach'],
                       c = over_50['target'],
                       cmap = "winter")
ax1.set(title = "thalach level",
        xlabel = "age",
        ylabel = "thalach",
        xlim = (50,80),
        ylim = [60,200])

ax1.legend(*scatter.legend_elements(), title = 'target')
ax1.axhline(y = over_50["thalach"].mean(),
            linestyle = '--')

fig.suptitle('Heart Disease Analysis', fontsize=12, fontweight='bold');
```
