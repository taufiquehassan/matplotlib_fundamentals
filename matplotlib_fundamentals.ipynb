### matplotlib workflow

```python
#0. import matplotlib
import matplotlib.pyplot as plt

#1. prepare data
x = [1,2,3,4,5]
y = [11,22,33,44,55]

#2. setup plot
fig, ax = plt.subplots(figsize = [10,10])

#3. plot data
ax.plot(x,y)

#4. customiz plot
ax.set(title = "simpole data",
       xlabel = "x-axis", 
       ylabel = "Yaxis")
      
#5. save and show
fig.savefig("sampleplot.png")
```

### Make the following plots:

* Line Plot
* Scatter Plot
* Bar Plot
* Histogram
* Subplot

```python
#plot the data
fig, ax = plt.subplots()
ax.plot(x,x**2);
```

```python
#plot the data scatter plot
fig, ax = plt.subplots()
ax.scatter(x,np.exp(x));
```

```python
#plot the data scatter plot
fig, ax = plt.subplots()
ax.scatter(x,np.sin(x))
```

```python
#Bar Plot
nut_butter_prices = {"Almond Butter" : 10,
                    "Peanut Butter": 8,
                    "Cashew butter": 20}

fig, ax = plt.subplots()
ax.bar(nut_butter_prices.keys(), nut_butter_prices.values());
ax.set(title = "Taufiques BUtter Prices",
      ylabel = "Price in ($)");
```

```python
#making horizontal bar graph
fig, ax = plt.subplots()
ax.barh(list(nut_butter_prices.keys()), list(nut_butter_prices.values()));
```

```python
x = np.random.randn(1000)
x
fig, ax = plt.subplots()
ax.hist(x)
```

```python
#Subplots
fig, ((ax1,ax2),(ax3,ax4)) = plt.subplots(nrows = 2,
                                          ncols = 2,
                                          figsize = (10,5))
ax1.plot(x,x/2)
ax2.bar(nut_butter_prices.keys(), nut_butter_prices.values())
ax3.scatter(np.random.random(10), np.random.random(10))
ax4.hist(x);
```
