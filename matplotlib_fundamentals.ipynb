### matplotlib workflow

```python
#0. import matplotlib
import matplotlib.pyplot as plt

#1. prepare data
x = [1,2,3,4,5]
y = [11,22,33,44,55]

#2. setup plot
fig, ax = plt.subplots(figsize = [10,10])

#3. plot data
ax.plot(x,y)

#4. customiz plot
ax.set(title = "simpole data",
       xlabel = "x-axis", 
       ylabel = "Yaxis")
      
#5. save and show
fig.savefig("sampleplot.png")
```

```python
#Bar Plot
nut_butter_prices = {"Almond Butter" : 10,
                    "Peanut Butter": 8,
                    "Cashew butter": 20}

fig, ax = plt.subplots()
ax.bar(nut_butter_prices.keys(), nut_butter_prices.values());
ax.set(title = "Taufiques BUtter Prices",
      ylabel = "Price in ($)");
```

```python
#making horizontal bar graph
fig, ax = plt.subplots()
ax.barh(list(nut_butter_prices.keys()), list(nut_butter_prices.values()));
```

```python
x = np.random.randn(1000)
x
fig, ax = plt.subplots()
ax.hist(x)
```

```python
#Subplots
fig, ((ax1,ax2),(ax3,ax4)) = plt.subplots(nrows = 2,
                                          ncols = 2,
                                          figsize = (10,5))
ax1.plot(x,x/2)
ax2.bar(nut_butter_prices.keys(), nut_butter_prices.values())
ax3.scatter(np.random.random(10), np.random.random(10))
ax4.hist(x);
```

## Plotting from Pandas dataframe

```python
car_sales = pd.read_csv('https://raw.githubusercontent.com/mrdbourke/zero-to-mastery-ml/master/data/car-sales.csv')
car_sales.head()
```

```python
fig, ax = plt.subplots()
ax.scatter(car_sales.Make, car_sales.Price)
```

```python
ts = pd.Series(np.random.randn(1000),
               index =  pd.date_range("1/1/2020", periods = 1000))
ts = ts.cumsum()
ts.plot();
```

```python
#replacing commas, dots and doller signs
car_sales["Price"] = car_sales["Price"].str.replace('[\$\,\.]', '')
car_sales["Price"] = car_sales["Price"].str[:-2]
car_sales
```

```python
car_sales["Sale Date"] = pd.date_range("1/1/2020", periods = len(car_sales))
car_sales["Total Sales"] = car_sales["Price"].astype(int).cumsum()
car_sales
```

```python
fig, ax = plt.subplots()
ax.plot(car_sales["Sale Date"], car_sales["Total Sales"]);
```

```python
#For Pandas DF
#car_sales.plot(x="Sale Date", y="Total Sales");
car_sales.plot("Sale Date", "Total Sales");
```

```python
car_sales["Price"] = car_sales["Price"].astype(int)
car_sales["Total Sales"] = car_sales["Total Sales"].astype(int)
#car_sales
car_sales.plot(x="Odometer (KM)", y="Price", kind = "scatter");
```
